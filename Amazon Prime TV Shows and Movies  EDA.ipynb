## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.


# 1. Removed Duplicated from titles dataset

titles = titles.drop_duplicates(keep = 'first')

# 2. Removed Duplicated from credits dataset

credits = credits.drop_duplicates(keep = 'first')

#### Handling Missing Values

# 3. filled NA in place of null values in column character 

credits.character = credits.character.fillna(value = 'NA')

# 4. filled NA in place of null values in column description 

titles.description = titles.description.fillna(value = 'NA')

# 5. filled NA in place of null values in column seasons 

titles.seasons = titles.seasons.fillna(value = 'NA')

# 6. filled NA in place of null values in column  age_certification

titles.age_certification = titles.age_certification.fillna(value = 'NA')

# 7. Removed Null values from column  Imdb Id

titles = titles.dropna(subset=['imdb_id'])

# 8. filled average value in place of null values in column imdb_score 

titles.imdb_score = titles.imdb_score.fillna(value = titles.imdb_score.mean())

# 9. filled median value in place of null values in column imdb_votes 

titles.imdb_votes = titles.imdb_votes.fillna(value=titles.imdb_votes.median())

# 10. filled average value in place of null values in column imdbtmdb_popularity

titles.tmdb_popularity = titles.tmdb_popularity.fillna(value=titles.tmdb_popularity.mean())

# 11. filled average value in place of null values in column tmdb_score 

titles.tmdb_score = titles.tmdb_score.fillna(value=titles.tmdb_score.mean())


# 12. merge both datasets as merged_data

merged_data = credits.merge(titles, on='id')


# 13. for converting in series

# import ast
#def convert_to_list(string):
#    return ast.literal_eval(string)
````

### What all manipulations have you done and insights you found?

1. Removed duplicate records from both titles and credits datasets.


2. Handled missing values:

* Filled missing character values with 'NA'.

* Filled missing description, seasons, and age_certification with 'NA'.

* Dropped null values from imdb_id.

* Replaced missing values in imdb_score, tmdb_score, and tmdb_popularity with the mean.

* Replaced missing values in imdb_votes with the median.



3. Merged titles and credits datasets using the id column.


4. Converted list-based columns like genres and production_countries into separate values for better analysis.

#### Chart - 4 - Distribution of IMDB Ratings with Before & After Average Rating

````python
# Chart - 4 visualization code

imdb_scores = titles['imdb_score'].dropna()

# Calculate Freedman-Diaconis bin width
q25, q75 = np.percentile(imdb_scores, [25, 75])
bin_width = 2 * (q75 - q25) * len(imdb_scores) ** (-1/3)
bins = int((imdb_scores.max() - imdb_scores.min()) / bin_width)

# Calculate mean and median
mean_score = imdb_scores.mean()
median_score = imdb_scores.median()

# Count values before and after mean
before_mean_count = (imdb_scores < mean_score).sum()
after_mean_count = (imdb_scores >= mean_score).sum()

# Set Seaborn style
sns.set_style("darkgrid")

# Create figure
plt.figure(figsize=(12, 6))

# Plot histogram with KDE
sns.histplot(imdb_scores, bins=bins, kde=True, color='royalblue', 
             edgecolor='black', alpha=0.6, linewidth=1.2)

# KDE curve customization
sns.kdeplot(imdb_scores, color='red', linestyle="--", linewidth=2)

# Add mean & median lines

plt.axvline(mean_score, color='green', linestyle='-', linewidth=2, label=f'Mean: {mean_score:.2f}')
plt.axvline(median_score, color='orange', linestyle='--', linewidth=2, label=f'Median: {median_score:.2f}')
plt.fill_betweenx(plt.ylim(), plt.xlim()[0], mean_score, color='blue', alpha=0.2, label=f'Before Mean: {before_mean_count}')
plt.fill_betweenx(plt.ylim(), mean_score, plt.xlim()[1], color='red', alpha=0.2, label=f'After Mean: {after_mean_count}')
plt.text(mean_score + 0.1, plt.ylim()[1] * 0.9, f'Mean: {mean_score:.2f}', color='green', fontsize=12, fontweight='bold')
plt.text(median_score - 0.6, plt.ylim()[1] * 0.8, f'Median: {median_score:.2f}', color='orange', fontsize=12, fontweight='bold')
plt.xlabel("IMDB Rating", fontsize=12, fontweight='bold')
plt.ylabel("Frequency", fontsize=12, fontweight='bold')
plt.title("Distribution of IMDB Ratings with Before & After Average Rating", fontsize=14, fontweight='bold')
plt.legend()
plt.show()
````

##### 1. Why did you pick the specific chart?

The histogram with a KDE (Kernel Density Estimate) was selected to show the distribution of IMDb ratings, allowing us to see the spread and density of ratings.

##### 2. What is/are the insight(s) found from the chart?

* Most ratings range from 5-7, indicating a large number of average-rated shows.

* Few extreme ratings (either very low or very high).
